## Usuwanie nie anglojęzycznych aplikacji
Każdy znak ma swój numer, sprawdza się go funkcją `ord()`
Korzystając z wiedzy że anglojęzyczny alfabet ma znaki w zakresie od 0 do 127.
W tym celu stworzona jest funkcja `czy_po_angielsku()`

```python
def czy_po_angielsku(string):

    for znak in string:
        if ord(znak) > 127:
            return False

    return True

print(czy_po_angielsku('Instagram'))
print(czy_po_angielsku('爱奇艺PPS -《欢乐颂2》电视剧热播'))
```

Należy pamiętać że niektóre aplikacje posaidają emotikony lub inne znaki spoza zakresu 0-127. Uznano, że aplikacja anglojęzyczna może posiadać 3 takei znaki.

Poniżej znajduje się ulepszona funkcja `czy_po_angielsku()`

```python
def czy_po_angielsku(string):
    poza_zakresem = 0

    for znak in string:
        if ord(znak) > 127:
            poza_zakresem += 1

    if poza_zakresem > 3:
        return False
    else:
        return True

print(czy_po_angielsku('Docs To Go™ Free Office Suite'))
print(czy_po_angielsku('Instachat 😜'))
```

Czas na zastosowanie funkcji na naszych zbiorach danych i stworzenie nowych, anglojęzycznych.

```python
android_po_angielsku = []
ios_po_angielsku = []

for app in android_czysty:
    nazwa = app[0]
    if czy_po_angielsku(nazwa):
        android_po_angielsku.append(app)

for app in ios:
    nazwa = app[1]
    if czy_po_angielsku(nazwa):
        ios_po_angielsku.append(app)

explore_data_pandas(android_po_angielsku, 0, 6)
print('\n')
explore_data_pandas(ios_po_angielsku, 0, 6)
```

## Najpolarniejsze Aplikacje ze względu na rodzaj.
Zajmijmy się stworzeniem funkcji które pomogą nam zobrazować częstość występowania danego typu aplikacji.

- pierwsza funkcja pokaże procentowy udział
- druga ułoży dane malejąco

```python
def tabela_frekwencji(dataset, index):
    tabela = {}
    w_sumie = 0

    for wiersz in dataset:
        w_sumie += 1 #aby potem podzielic przez calosc
        wartosc = wiersz[index]
        if wartosc in tabela:
            tabela[wartosc] += 1 #jezeli jest to dodaj 1 do wartosci
        else:
            tabela[wartosc] = 1 #jezeli nie ma to wartosc = 1

    tabela_procentowa = {}
    for klucz in tabela:
        procent = (tabela[klucz] / w_sumie) * 100 #ilosc dodanych przez wszytskie
        tabela_procentowa[klucz] = procent #

    return tabela_procentowa


def wyswietl_tabele(dataset, index):
    tabela = tabela_frekwencji(dataset, index)
    tw = [] #pusta lista
    for klucz in tabela: #klucz to nazwa aplikacji
        klucz_wartosc = (tabela[klucz], klucz) #
        tw.append(klucz_wartosc)

    posortowana_tabela = sorted(tw, reverse = True)
    for wartosc in posortowana_tabela:
        print(wartosc[1], ':', wartosc[0])
```

```python
wyswietl_tabele(ios_darmowe, -5) #kolumna prime_genre IOS
```

```python
wyswietl_tabele(android_darmowe, 1) #kolumna category Android
```

```python
wyswietl_tabele(android_darmowe, -4) #kolumna genres
```

Populaność aplikacji czyli to ile ma ona uzytkownikow sprawdzic mozna liczbą instalacji, w zbiorze danych AppStore nie ma takiej informacji, użyta zostanie kolumna `rating_count_tot` w któ®ej znajdują się infoamcje o całkowitej liczbie ocen danej aplikacji.

### App Store


```python
oceny_total_ios = tabela_frekwencji(ios_darmowe, -5) # -5 to prime_genre
#print('>>>')
#print(oceny_total_ios)
for rodzaj in oceny_total_ios:
  #
    razem = 0
    dlugosc_r = 0
    for app in ios_darmowe:
        rodzaj_apki = app[-5]
        if rodzaj_apki == rodzaj:
            n_ocen = float(app[5]) # 5 to ilosc ocen
            razem += n_ocen #za kazdym przejsciem pętli gdy znajdzie się
            dlugosc_r += 1
    #print('razem:')
    #print(razem)
    #print('dlugosc r:')
    #rint(dlugosc_r)
    sr_n_ocen = razem / dlugosc_r
    print(rodzaj, ':', sr_n_ocen)
```

### Android
W zbiorze danych istanieje kolumna `Installs`

```python
wyswietl_tabele(android_darmowe, 5)
```

W danych jest dość spora dysproporcja, rangi rosną wykładniczo i ciężko o miarodajność, ponizej podwójną funckją `for` obliczono średnią ilość instalacji dla danej kategorii aplikacji.

```python
kategorie_android = tabela_frekwencji(android_darmowe, 1)

for kategoria in kategorie_android:
    razem = 0
    dl_zbioru_kategorii = 0
    for app in android_darmowe:
        categoria_apki = app[1]
        if categoria_apki == kategoria:
            n_istalacji = app[5]
            n_istalacji = n_istalacji.replace(',', '')
            n_istalacji = n_istalacji.replace('+', '')
            razem += float(n_istalacji)
            dl_zbioru_kategorii += 1
    sr_n_instalacji = razem / dl_zbioru_kategorii
    print(kategoria, ':', sr_n_instalacji)
```

Kategoria Communication posiada największą srednią liczbe instalacji, jest ona zawyzona przez takie giganty jak Waze czy Google Maps.
